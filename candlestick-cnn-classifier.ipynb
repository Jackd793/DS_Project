```python
from mpl_finance import candlestick_ohlc
from numpy import random
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Creating Patterns

```python
df = pd.DataFrame()
while len(df) < 1000:
  try:
    start = random.randint(0, high=100)
    close_yest = random.randint(0, high=start)
    open_yest = random.randint(close_yest, high=100)
    yest_low = random.randint(0, high=close_yest)
    yest_high = random.randint(open_yest, high=100)

    close_today = random.randint(open_yest, high=100)
    open_today = random.randint(0, high=close_yest)
    today_low = random.randint(0, high=open_today)
    today_high = random.randint(close_today, high=100)

    # new_row = {'close_yest': close_yest, 'open_yest': open_yest, 'close_today': close_today, 'open_today': open_today}
    new_row1 = {'date': 1, 'open': open_yest, 'high': yest_high, 'low': yest_low,'close': close_yest}
    new_row2 = {'date': 2, 'open': open_today, 'high': today_high, 'low': today_low,'close': close_today}

    df = df.append(new_row1, ignore_index = True)
    df = df.append(new_row2, ignore_index = True)

    # re-order columns
    df = df[['date','open','high','low','close']]

  except:
    pass
# View df
# df


pattern_dict = {}
j=1
lst_bullish_engulfing = []

i= 0
for i in range(0,len(df)-2,2):
  print(i)
  two_day_chart = df[i:i+2].copy()

  low = two_day_chart['low'].min()
  high = two_day_chart['high'].max()

  # Normalise each chart to range from 0 to 3 as we are looking at a 3 day chart
  # This prduces square matplotlib charts
  two_day_chart[['open','high','low','close']] = two_day_chart[['open','high','low','close']].apply(lambda x : ((x-low)/(high-low))*2)

  fig, ax = plt.subplots()
  candlestick_ohlc(ax, two_day_chart[:2].values, width=1, colorup='g', colordown='r');

  ax.axis('off')

  plt.axis('square')
  plt.show()

  Z = np.array(fig.canvas.renderer._renderer)
  lst_bullish_engulfing.append(Z)
  j+=1

  pattern_dict['bullish_engulfing_'+str(j)] = Z
```

# 3Day Patterns

```python
# Evening Star
# 3 day pattern

df = pd.DataFrame()
while len(df) < 1000:
  try:
    start = random.randint(0, 50)
    # uptrend
    day1_open = random.randint(0, 50)
    day1_close = random.randint(day1_open, 50)
    day1_high = random.randint(day1_close,75)
    day1_low = random.randint(0,day1_open)

    day2_close = random.randint(day1_close,75)
    day2_open = random.randint(day2_close,100)
    # high price and opening price distance must be twice as large as the body
    day2_high = random.randint(day2_close,100)
    # little or no lower shadow
    day2_low = random.randint(50,day2_close*0.9)

    # confirmation of reversal
    day3_open = random.randint(50,day2_close)
    day3_close = random.randint(0,day3_open)
    day3_high = random.randint(day3_open,day3_open*1.1)
    day3_low = random.randint(0,day2_close)

    new_row1 = {'date': 1, 'open': day1_open, 'high': day1_high, 'low': day1_low,'close': day1_close}
    new_row2 = {'date': 2, 'open': day2_open, 'high': day2_high, 'low': day2_low,'close': day2_close}
    new_row3 = {'date': 3, 'open': day3_open, 'high': day3_high, 'low': day3_low,'close': day3_close}


    df = df.append(new_row1, ignore_index = True)
    df = df.append(new_row2, ignore_index = True)
    df = df.append(new_row3, ignore_index = True)


    # re-order columns
    df = df[['date','open','high','low','close']]

  except:
    pass
```

```python
lst_shooting_star = []
i= 0
j=0
for i in range(0,len(df)-3,3):
  print(i)
  three_day_chart = df[i:i+3].copy()

  low = three_day_chart['low'].min()
  high = three_day_chart['high'].max()

  # Normalise each chart to range from 0 to 3 as we are looking at a 3 day chart
  # This prduces square matplotlib charts
  three_day_chart[['open','high','low','close']] = three_day_chart[['open','high','low','close']].apply(lambda x : ((x-low)/(high-low))*3)

  fig, ax = plt.subplots()
  candlestick_ohlc(ax, three_day_chart[:3].values, width=1, colorup='g', colordown='r');

  ax.axis('off')

  plt.axis('square')
  plt.show()

  Z = np.array(fig.canvas.renderer._renderer)
  lst_shooting_star.append(Z)
  j+=1

  pattern_dict['shooting_star_'+str(j)] = Z
```

```python
Xtrain = np.array(list(pattern_dict.values()))
Xtrain.shape
```

```python
Ytrain = [0]*499+[1]*499+[2]*499
```

```python
Xtrain = Xtrain.astype('float32')
Xtrain /= 255
```

```python
from keras.utils import np_utils
```

```python
# Classify as categorical as one hot vector matrix
Ytrain = np_utils.to_categorical(Ytrain,3)
Ytrain[0]
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten, Convolution2D, MaxPooling2D

from keras.datasets import mnist
```

```python
model = Sequential()
# Add layer
model.add(Convolution2D(32,(3,3),input_shape=Xtrain[0].shape))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))
# Turn data into 1D data from 2D
model.add(Flatten())
model.add(Dense(32))
model.add(Dense(3,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(Xtrain,Ytrain,batch_size=12,epochs=5,verbose=1,validation_split=0.1)
```

```python
model.predict()
```
